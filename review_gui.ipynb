```python
import tkinter as tk
from tkinter import *
import tkinter.font as font
from PIL import Image,ImageTk
import string
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import WhitespaceTokenizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

import pickle
model=pickle.load(open('RestReviewmodel','rb'))

def clean_text(text):
    # lower text
    text = text.lower()
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # remove words that contain numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    # remove stop words
    stop = stopwords.words('english')
    text = [x for x in text if x not in stop]
    # remove empty tokens
    text = [t for t in text if len(t) > 0]
    # remove words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
    text = " ".join(text)
    return(text)

data["review_clean"] = data["reviews"].apply(lambda x: clean_text(x))

def rate():
    review=text.get("1.0","end-1c")
    l1 = clean_text(review)
    l=[l1]
    wordvector = CountVectorizer()
    finalwordvector = wordvector.fit(data['review_clean'])
    l2 = finalwordvector.transform(l)
    tfidf = TfidfTransformer().fit(l2)
    l3 = tfidf.transform(l2)
    print(l3)
    a=l3[0]
    review_rate = model.predict(a)
    print(review_rate)
    message = "Your rate is: "+str(review_rate[0])
    messagebox.showinfo("Review Rating",message)

top = Tk()

top.geometry("660x480")

top.title("RS")

top.columnconfigure(0, pad=3)
top.columnconfigure(1, pad=3)
top.columnconfigure(2, pad=3)
top.columnconfigure(3, pad=3)

top.rowconfigure(0, pad=3)
top.rowconfigure(1, pad=3)
top.rowconfigure(2, pad=3)
top.rowconfigure(3, pad=3)
top.rowconfigure(4, pad=3)
top.rowconfigure(5, pad=3)
top.rowconfigure(6, pad=3)
#top.rowconfigure(7, pad=3)

lbl = Label(top, text = " Abstract")
lbl.grid(row=0,column=0, columnspan=4)

T = Text(top,height=14, width=80)
T.grid(row=1,columnspan=2, padx=(5,5),pady=(5,5))

Fact = """In our work we have develop an overall process of restaurant review based sentiment analysis.
Variety of research has been performed in opinion mining.
Sentiment analysis has ever been a problem to identify with several type of data whether it is for products or services, restaurant review. 
Different user given rating reviews based feature selection can mislead to improper and inaccurate results.
The challenge here is to detect of spam contents in usersâ€™ review and to apply the study of this problem to solve. 
Improving accuracy and finding positive and negative review about any restaurant reviews.
To use the database and opinion mining process by sentiment analysis"""

lbl = Label(top, text = " Type your review here")
lbl.grid(row=2,column=0, columnspan=4)

text = Text(top, height=8, width=80)
text.grid(row=3, rowspan=3, columnspan=2, padx=(5,5),pady=(5,5))

sub = Button(top, text="Is the given review negative or positive",command=rate)
sub.grid(row=6,column = 0, columnspan=4)

T.insert(tk.END, Fact)

top.mainloop()
```
